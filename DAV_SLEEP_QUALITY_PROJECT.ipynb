**Handling the Missing Values**

```python
missing_df = data.isna().sum()
missing_df_data = pd.DataFrame(missing_df).reset_index()
missing_df_data.columns = ['Column', 'Missing']
missing_df_data
```

**DATA PRE**-**PROCESSING**

```python
from sklearn import preprocessing

df = data.copy()
label_encoder = preprocessing.LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['Occupation'] = label_encoder.fit_transform(df['Occupation'])
df['BMI Category'] = label_encoder.fit_transform(df['BMI Category'])
df['Sleep Disorder'] = label_encoder.fit_transform(df['Sleep Disorder'])
df.drop('Blood Pressure', axis = 1, inplace = True)
df.head()
```

**FINDING THE RELATIONSHIPS BETWEEN THE ATTRIBUTES**

**1. Relationship between sleep duration and sleep quality**

```python
Correlation = data['Sleep Duration'].corr(data['Quality of Sleep'])
print('relationship between sleep duration and sleep quality: ', Correlation)

plt.scatter(data['Sleep Duration'], data['Quality of Sleep'])
plt.xlabel('Sleep duration')
plt.ylabel('Quality of sleep')
plt.title('relationship between sleep duration and sleep quality')
plt.show()
```

Inference: First I saw the correlation between sleep duration and sleep quality, it turns out that the correlation shows the number 0.88, this is a fairly high correlation number. while on the graph it can be read that people who have long sleep duration tend to have good sleep quality

***2. The Relationship between Physical Activity and Sleep Quality ***

```python
sns.boxplot(x='Physical Activity Level', y='Quality of Sleep', data=data)
plt.xlabel('Physical activity level')
plt.ylabel('Quality of sleep')
plt.title('The Relationship between Physical Activity and Sleep Quality')
plt.show()
```

```python
plt.figure(figsize = (20, 10))
plt.subplot(2, 1, 1)
plt.gca().set_title('Variable Gender')
sns.countplot(x = 'Gender', hue = 'Sleep Disorder', data = data)

plt.subplot(2, 1, 2)
plt.gca().set_title('Variable BMI Category')
sns.countplot(x = 'BMI Category', hue = 'Sleep Disorder', data = data)
```

**Problem Statement : Plot the histogram for the count of the people having the Sleep Apnea,Insomnia,& None**

```python
import plotly.express as px
fig=px.histogram(data,x='Sleep Disorder', barmode="group", color='Sleep Disorder', text_auto=True)
fig.update_yaxes(showgrid=False)
fig.update_layout(title='<b>Sleep Disorder Counts</b>', title_font={'size':20})

fig.show()
```

**Problem Statement : How many Males and Females are suffering with Insomnia ,Sleep Apnea and how many are happily sleeping ?**

```python
data.groupby('Sleep Disorder')['Gender'].value_counts()
```

**Problem Statement : Plot the Tree Map by taking 'Jobs','Sleep Disorder','Occupation'  Attributes.**

```python
fig=px.treemap(data,path=[px.Constant('Jobs'),'Sleep Disorder','Occupation'], color='Sleep Disorder')
fig.update_layout(title='<b>Occupation - Sleep Disorder</b>', title_font={'size':20})

fig.show()
```

**Problem Statement: Plot the Sunburst Display by taking 'Sleep quality' , 'Sleep Disorder' , 'Quality of Sleep' Attributes.**

```python
fig=px.sunburst(data,path=[px.Constant('Sleep quality'),'Sleep Disorder','Quality of Sleep'],
               color='Sleep Disorder',values='Sleep Duration', hover_data=['Gender'])
fig.update_layout(title='<b>Quality of Sleep - Sleep Disorder</b>', title_font={'size':20})

fig.show()
```

**Problem Statement: Plot the Violin Plot by taking 'Sleep Disorder' , 'Physical Activity Level' Attributes.**

```python
fig = px.violin(data, x="Sleep Disorder",y='Physical Activity Level',color='Sleep Disorder', violinmode='overlay')
fig.update_yaxes(showgrid=False)
fig.update_layout(title='<b>Physical Activity Level - Sleep Disorder</b>', title_font={'size':20})

fig.show()
```

**Problem Statement : Plot the graph for probability of getting the Sleeping Disorder considering Age.**

**Using Empirical distribution function.**

```python
fig=px.ecdf(data,x='Age', color='Sleep Disorder')
fig.update_layout(title='<b>Age - Sleep Disorder </b>', title_font={'size':20})
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)

fig.show()
```
